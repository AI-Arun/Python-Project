                                              #1.Assignment-Visualization-titanic data set

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the data source
df=pd.read_csv(r'titanic_original.csv')
df
```

```python
df.tail()
```

```python
df["sex"].isnull().sum()
```

```python
df.dropna(subset=['sex'], inplace=True)
```

```python
df['sex'].unique()
```

```python
#1.Create a pie chart presenting the male/female proportion 
df['sex'].value_counts().plot.pie(autopct='%1.0f%%',title='Distribution of Sex')
plt.ylabel('')
plt.show()
```

```python
#2.Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
plt.figure(figsize=(10, 6))
sns.relplot(x= "age",y= "fare",data = df, hue = "sex")
plt.title('Scatter Plot of Fare Paid vs Age')
```

                                          #2.Assignment-Visualization- Diamonds data set

Diamonds_url.CSV file containing various attributes like carat, cut, color, clarity, price etc. for over 53,000 diamonds. 
Visualize the relationship between the carat (size of diamond) and price using a scatter plot. 
Instead of using the entire dataset for this visualization, just pick the diamonds with a clarity "SI2" and color "E". 
Use the values of the "cut" column to color the dots in the scatter plot. 
Make appropriate modifications to the chart title, axis titles, legend, figure size, font size, colors etc. 
To make the chart readable and visually appealing.

```python
df1 = pd.read_csv(r'diamonds.csv')
df1
```

```python
df1.info()
```

```python
df_filtered = df1[(df1['clarity'] == 'SI2') & (df1['color'] == 'E')]
```

```python
df_filtered
```

```python
sns.relplot(x=df_filtered['carat'], y=df_filtered['price'],hue=df_filtered['cut'])
plt.title('Price of the Diamonds by Carat (Clarity: SI1) And colour:E',fontsize=16)
plt.xlabel('Carat',fontsize=14)
plt.ylabel('Price',fontsize=14)
plt.show()
```

                                                #3.Assignment-Visualization-Planet data set

The Planets dataset contains details about the 1,000+ extrasolar planets discovered up to 2014. 
Visualize the distribution of the masses of the planets (expressed as a multiple of the mass of Jupiter), 
using a histogram and a box plot. Make appropriate modifications to the chart title, axis titles, legend, figure size, font size, colors etc. 
to make the chart readable and visually appealing.

```python
df2= pd.read_csv(r'planets.csv')
df2
```

```python
df2.info()
```

```python
masses = df2['mass'].dropna()
```

```python
masses
```

```python
fig,ax=plt.subplots(2,1,figsize=(10,12)) 
ax[0].hist(masses,bins=30, color='skyblue', edgecolor='black')
ax[0].set_title('Distribution of Planet Masses (Multiple of Jupiter\'s Mass)', fontsize=18)
ax[0].set_xlabel('Mass (Jupiter Masses)', fontsize=14)
ax[0].set_ylabel('Frequency', fontsize=14)
ax[0].tick_params(axis='both', labelsize=12)

ax[1].boxplot(masses,vert=False,patch_artist=True, boxprops=dict(facecolor='skyblue'))
ax[1].set_title('Box plot of Planet Masses (Multiple of Jupiter\'s Mass)', fontsize=18)
ax[1].set_xlabel('Mass (Jupiter Masses)', fontsize=14)
ax[1].set_yticklabels([''])
ax[1].tick_params(axis='x', labelsize=12)
```
